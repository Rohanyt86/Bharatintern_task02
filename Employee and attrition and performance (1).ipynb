#### Comparing Distance from home with attrition

```python
#  compare the average "DistanceFromHome" for employees who left and those who stayed

distance_from_home_by_attrition = df.groupby("Attrition")["DistanceFromHome"].mean()
print("Distance From Home by Attrition:")
print(distance_from_home_by_attrition)
```

```python
## above data use in graph

plt.figure(figsize=(5,5))
plt.bar(list(df.groupby("Attrition")["DistanceFromHome"].mean().keys()),df.groupby("Attrition")["DistanceFromHome"].mean(),color = ["b","g"])
plt.xlabel("Attrition")
plt.ylabel("Average Distance from Home")
plt.title("Distance from home by Attrition",size=10,color = 'g')
plt.show()
```

we can interpret that employees who stayed at the company have an average distance of approximately 8.92 units 
from their homes to the workplace. On the other hand, employees who left the company have an average distance of 
approximately 10.63 units from their homes to the workplace. This indicates that, on average, 
employees who left the company tended to have a slightly longer commuting distance compared to those who stayed.

### comparing  MonthlyIncome with Attrition

```python
MonthlyIncome_Attrition = df.groupby("Attrition")["MonthlyIncome"].mean()
print("MonthlyIncome_Attrition:")
print(MonthlyIncome_Attrition)
```

```python
## above details use histplot graph

# Filter data for employees who stayed (Attrition = "No") & who left (Attrition = "Yes")
monthly_income_stayed = df[df["Attrition"] == "No"]["MonthlyIncome"]
monthly_income_left = df[df["Attrition"] == "Yes"]["MonthlyIncome"]

# Set up the figure and axes for the plot
plt.figure(figsize=(8, 6))

# Create the histogram for MonthlyIncome of employees who stayed
sns.histplot(monthly_income_stayed, color='blue', label='Stayed', kde=True)

# Create the histogram for MonthlyIncome of employees who left
sns.histplot(monthly_income_left, color='red', label='Left', kde=True)

# Customize the plot
plt.xlabel("Monthly Income")
plt.ylabel("Frequency")
plt.title("Monthly Income by Attrition")
plt.legend()
plt.show()

```

Histogram can easily tells that Average of Monthly income is not responsible for Employee left.

#### Statistics 

```python
from scipy.stats import ttest_ind

## T-test and Hypothesis Testing
monthly_income_stayed = df[df["Attrition"] == "No"]["MonthlyIncome"]
monthly_income_left = df[df["Attrition"] == "Yes"]["MonthlyIncome"]

t_stat, p_value = ttest_ind(monthly_income_stayed, monthly_income_left)
alpha = 0.05

print("T-Test Results:")
print("T-statistic:", t_stat)
print("P-value:", p_value)

if p_value < alpha:
    print("The difference in mean monthly income between employees who stayed and those who left is statistically significant.")
else:
    print("There is no statistically significant difference in mean monthly income between employees who stayed and those who left.")
```
